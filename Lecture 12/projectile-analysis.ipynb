```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('projectile_data.csv')
print(df.head())
```

```python
df_with_velocity = df
time_diff = [current - previous for current, previous in zip(df['time'][1:], df['time'])]
pos_diff = [current - previous for current, previous in zip(df['x_position'][1:], df['x_position'])]
vel_x = [pos_diff[i] / time_diff[i] for i in range(len(time_diff))]
df_with_velocity['velocity_x'] = [0] + vel_x
print(df_with_velocity)
```

```python
df_with_velocity = df
#time_diff = [current - previous for current, previous in zip(df['time'][1:], df['time'])]
pos_diff = [current - previous for current, previous in zip(df['y_position'][1:], df['y_position'])]
vel_y = [pos_diff[i] / time_diff[i] for i in range(len(time_diff))]
df_with_velocity['velocity_y'] = [0] + vel_y
print(df_with_velocity)
```

```python
df_with_mag_velocity = df_with_velocity
import numpy as np
mag = [comp_x**2 + comp_y**2 for comp_x, comp_y in zip(df['velocity_x'], df['velocity_y'])]
df_with_mag_velocity['velocity_magnitude'] = np.sqrt(mag)
print(df_with_mag_velocity)
```

```python
max_height_Index = df_with_mag_velocity['y_position'].idxmax()
print(max_height_Index)
max_height_time = df_with_mag_velocity.loc[max_height_Index, 'time']
max_height = df_with_mag_velocity.loc[max_height_Index, 'y_position']

print(f"The maximum height is at {max_height} at time = {max_height_time}")
```

```python
# Exclude the launch point by slicing the DataFrame
df_excluding_launch = df_with_mag_velocity[1:]
gnd_Index = df_excluding_launch['y_position'].idxmin()
gnd_HitTime = df_with_mag_velocity.loc[gnd_Index, 'time']
print(f"The time of impact to the ground is {gnd_HitTime}")
```

```python
df_with_mag_velocity.plot()
plt.xlabel('time')
plt.show()
```
