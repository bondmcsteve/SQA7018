```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Reading the data into DataFrame

```python
df = pd.read_csv("spectrum_data.csv")
print(df.head())
print(df.shape)
```

### Finding wavelength with peak intensity

```python
max_Intensity_Index = df['intensity'].idxmax()
print(max_Intensity_Index)
max_Intensity = df.loc[max_Intensity_Index, 'intensity']
max_Intensity_freq = df.loc[max_Intensity_Index, 'wavelength']

print(f"The maximum radiation intensity is at {max_Intensity:.6g} at wavelength = {max_Intensity_freq:.9g}")
```

### Calculating Wien's Displacement Law for Blackbody Temperature

Given that the Wien's displacement law is:

$ \lambda_{max}T = b $

```python
b = 2.897 * 10**6  #in nm*K
T = [b / photon_len for photon_len in df['wavelength']]
df_with_Wien = df
df_with_Wien['blackbody temperature Wien Law'] = T
df_with_Wien.head()
```
